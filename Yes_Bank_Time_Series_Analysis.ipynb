### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Stetement: The Average Opening Price of the Yes Bank Stocks is 100 having a standard deviation of 99.

Null Hyposthesis: H0: Mean = 100

Alternate Hypothesis HA: Mean != 100

Lets Assume the values of alpha be 0.05

Let the sample size be 50

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Given information
sample_mean = 100
population_mean = 105.5414054054054
population_std = 99
sample_size = 50
alpha = 0.05

# compute the z-score
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(50))
print('Z-Score :',z_score)

# Approach 1: Using Critical Z-Score

# Critical Z-Score
z_critical = stats.norm.ppf(1-alpha)
print('Critical Z-Score :',z_critical)

# Hypothesis
if z_score >  z_critical:
    print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")

# Approach 2: Using P-value

# P-Value : Probability of getting less than a Z-score
p_value = 2*(1-stats.norm.cdf(z_score))

print('p-value :',p_value)

# Hypothesis
if p_value <  alpha:
    print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Z Statistic Test has been done to Calculate P-Value.

##### Why did you choose the specific statistical test?

Since The Value of standard deviation was assumed and the sample size was greater than 30 we chose this test.

````python
# Installing pmdarima library
!pip install pmdarima

# Importing Necessary Libraries
import os
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
dateparse = lambda dates: pd.datetime.strptime(dates, '%b-%d')
stock_data = pd.read_csv('/content/Yes Bank Stock Price.csv', parse_dates=['Date'], date_parser=dateparse).fillna(0)
stock_data
````

````python
#Distribution of the dataset
df_close = stock_data['Close']
df_close.plot(kind='kde')
````

````python
#Test for staionarity
def test_stationarity(timeseries):
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    #Plot rolling statistics:
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
test_stationarity(df_close)
````

The graph above shows an increasing mean and standard deviation, suggesting that our series is not stationary.

The p-value is greater than 0.05, and the test statistics exceed the critical values, which means we cannot reject the Null hypothesis. Moreover, these findings indicate that the data is nonlinear.

Before conducting a time series analysis, it may be necessary to isolate the seasonality and trend components from our data series. This separation process can potentially lead to a stagnant resulting series.

### Splitting Dataset into Train and Test

````python
# Splitting data into train and training set
train_data, test_data = df_log[:int(len(df_log)*0.9)], df_log[int(len(df_log)*0.9):]

# Displaying Traing Dataset and Testing Dataset
plt.figure(figsize=(10,6))
plt.grid(True)
plt.xlabel('Dates')
plt.ylabel('Closing Prices')
plt.plot(df_log, 'green', label='Train data')
plt.plot(test_data, 'blue', label='Test data')
plt.legend()
````

Using auto_arima() function finding the best suitable hyperparameters p,q,d for traing the ARIMA model instead of using ACF and PACF Charts.

````python
model_autoARIMA = auto_arima(train_data, start_p=0, start_q=0,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=1,              # frequency of series
                      d=None,           # let model determine 'd'
                      seasonal=False,   # No Seasonality
                      start_P=0,
                      D=0,
                      trace=True,
                      error_action='ignore',
                      suppress_warnings=True,
                      stepwise=True)
print(model_autoARIMA.summary())
model_autoARIMA.plot_diagnostics(figsize=(15,8))
plt.show()
````

````python
from statsmodels.tsa.arima.model import ARIMA

# Training ARIMA Model with hyperparameters 1,1,1 i.e. p,d,q
model = ARIMA(train_data, order=(1,1,1))
fitted = model.fit()
print(fitted.summary())
````

````python
# Forecast
fc = fitted.forecast(500,alpha=0.05)  # with 95% confidence

# Make as pandas series
fc_series = pd.Series(fc, index=test_data.index)
# Plot
plt.figure(figsize=(10,5), dpi=100)
plt.plot(train_data, label='training data')
plt.plot(test_data, color = 'blue', label='Actual Stock Price')
plt.plot(fc_series, color = 'orange',label='Predicted Stock Price')
plt.title('YES BANK Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('YES BANK Stock Price')
plt.legend(loc='upper left', fontsize=8)
plt.show()
````
